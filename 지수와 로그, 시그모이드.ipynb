```python
# 왜.. 이걸 알아야 하죠?

# 우리가 지수 및 로그 함수를 배우는 이유는 통계 프로그래밍이나 예측 그리고 머신러닝의 정확도를 가늠할 수 있는 척도를 만들기 위해서이다.
# 지수 함수는 대표적으로 딥러닝에서 최적화를 위해 사용하는 활성화 함수인 '소프트맥스' 라는 라이브러리에 아주 적절하게 사용됩니다.

# 대표적인 척도가 '가능도(likelihood)' 입니다.
# 가능도의 척도를 지닌 정확도 측정 함수들은 0 ~ 1 값을 가지고 있습니다. 머신러닝으로 계속된 연산을 하게 되면 값이 어마어마하게 커집니다.
# 그렇기에 1 이하 값을 이용하는 로그 함수를 이용해야 계산이 빠르고 결과적으로 답을 빨리 얻을 수 있습니다.

# 역 함수 관계 y=x^2 <=> log_Y=x
# 둘 다, 큰 값을 작게 줄여 활용하기에 탄생된 수학 개념

import numpy as np
import matplotlib.pyplot as plt


values = np.array([ -10, -3, -1, 0.00003, 0.5, 1, 2, 4, 5, 8, 11, 13, 15 ])
```

```python
# 지수 함수 (지수로그 사용)
exp_values = np.exp(values)
print("지수 :", exp_values)

plt.plot(values, exp_values, color="red", linewidth=3,marker=".", ms=5, mec="green", mew=5, mfc="blue")
plt.title("func exp")
plt.xlabel("count")
plt.ylabel("exp values")
plt.show()
```

```python
# 로그 함수 (자연로그 사용)
log_values = np.log(values)
print("로그 :", log_values)

plt.plot(values, log_values, color="red", linewidth=3,marker=".", ms=5, mec="green", mew=5, mfc="blue")
plt.title("func log")
plt.xlabel("count")
plt.ylabel("log values")
plt.show()
```

```python
# 시그모이드 함수
sigmoid_values = 1 / ( 1 + np.exp( -values ) )
print("시그모이드 :", sigmoid_values)

plt.plot(values, sigmoid_values, color="red", linewidth=3,marker=".", ms=5, mec="green", mew=5, mfc="blue")
plt.title("func sigmoid")
plt.xlabel("count")
plt.ylabel("sigmoid values")
plt.grid(True)
plt.show()
```
